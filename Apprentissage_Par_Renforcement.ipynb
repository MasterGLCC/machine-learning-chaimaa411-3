## Q-learning : 

```python
import random  # importer le module random pour faire des choix aléatoires

# 1. Définition des états et actions possibles
states = ["A", "B"]   # A = position actuelle, B = autre position(état final)
actions = ["gauche", "droite"]

# 2. Initialisation de la Q-table (au début tout vaut 0)
# chaque état possède :  une valeur pour gauche et une valeur pour droite
Q = {
    s: {a: 0 for a in actions} 
    for s in states
}

# 3. Paramètres d'apprentissage
alpha = 0.1      # vitesse d'apprentissage 
gamma = 0.9      # importance des récompenses futures
epsilon = 0.3    # probabilité d'exploration, ici 0.3 = 30% d'exploration aléatoire
episodes = 10    # nombre de parties d'entraînement

# 4. Environnement
# Cette fonction définit les règles du monde, elle retourne la récompence et le prochain état 
def env(state, action):

    # si on est en A et qu'on va à droite -> récompense 10, on arrive en B
    if state == "A" and action == "droite":
        return 10, "B"   

    # si on est en A et qu'on va à gauche -> récompense 0, on reste en A
    if state == "A" and action == "gauche":
        return 0, "A"
        
    # si on est en B (état terminal) -> récompense 0, on reste en B
    if state == "B":
        return 0, "B"   # plus rien à apprendre
        
    return 0, state  # sécurité si aucun cas trouvé

# 5. Boucle d'apprentissage (répétée pour chaque épisode)
for episode in range(episodes):
    
    state = "A"   # on commence toujours en A au début d'un épisode
    
    # Tant qu'on n'a pas atteint l'état B (objectif), on continue
    while state != "B":
        
        if random.uniform(0, 1) < epsilon:   # nombre aléatoire entre 0 et 1
            action = random.choice(actions)  # choisir la meilleure valeur Q

        else:
            action = max(Q[state], key=Q[state].get)  # choisir l’action ayant la plus grande valeur Q
            
        reward, next_state = env(state, action)   # exécuter l’action et récupérer la récompense et le nouvel état
        old_q = Q[state][action]   # sauvegarder l’ancienne valeur

        # si le prochain état n’est pas B
        if next_state != "B":  
            max_next = max(Q[next_state].values())     # prendre la meilleure valeur future
        else:
            max_next = 0  # si état final aucune récompense future
            
        # Formule de mise à jour du Q-learning : Q(s,a) <- Q(s,a) + α * [ récompense + γ * max(Q(s',a')) - Q(s,a) ]
        new_q = old_q + alpha * (reward + gamma * max_next - old_q)
        
        Q[state][action] = new_q    # On enregistre la nouvelle valeur dans la Q-table
        state = next_state    # On passe à l'état suivant pour la prochaine itération

# 6. Affichage du résultat final
print("Q-table après apprentissage :")   # afficher le contenu final de la Q-table
for s in states:  # parcourir tous les états
    print(f"État {s} : {Q[s]}")   # afficher les valeurs Q de chaque état
```
