## Naive Bayes : 

```python
# 1. Données d'entraînement (Temps, Vent, Classe)
data = [

    ("Soleil", "Faible", "Oui"),
    ("Soleil", "Fort", "Non"),
    ("Pluie", "Faible", "Oui"),
    ("Pluie", "Fort", "Non")

]

# 2. Exemple à prédire On veut savoir : la classe est "Oui" ou "Non" ?
temps = "Soleil"
vent = "Faible"

# 3. Compter les classes
oui = 0   # nombre des "Oui"
non = 0   # nombre des "Non"

# parcourir les données
for d in data:
    if d[2] == "Oui":   # d[2] = classe
        oui += 1

    else:
        non += 1

# 4. Probabilités des classes
p_oui = oui / len(data)   # P(Oui)
p_non = non / len(data)   # P(Non)

# 5. Probabilités conditionnelles

# P(Soleil | Oui) : Parmi les lignes "Oui" combien ont Soleil ?
soleil_oui = 0
for d in data:

    if d[0] == "Soleil" and d[2] == "Oui":
        soleil_oui += 1

p_soleil_oui = soleil_oui / oui

# P(Faible | Oui) : Parmi les lignes "Oui" combien ont Faible ?
faible_oui = 0
for d in data:

    if d[1] == "Faible" and d[2] == "Oui":
        faible_oui += 1

p_faible_oui = faible_oui / oui

# P(Soleil | Non) : Parmi les lignes "Non" combien ont Soleil ?
soleil_non = 0
for d in data:

    if d[0] == "Soleil" and d[2] == "Non":
        soleil_non += 1

p_soleil_non = soleil_non / non

# P(Faible | Non)  : Parmi les lignes "Non" combien ont Faible ?
faible_non = 0
for d in data:

    if d[1] == "Faible" and d[2] == "Non":
        faible_non += 1

p_faible_non = faible_non /non

# 6. Calcul score pour "Oui"
# P(Oui) × P(Soleil|Oui) × P(Faible|Oui)
score_oui = (p_oui * p_soleil_oui *  p_faible_oui)

# 7. Calcul score pour "Non"
# P(Non) × P(Soleil|Non) × P(Faible|Non)
score_non = (p_non * p_soleil_non * p_faible_non)

# 8. Affichage des scores
print("Score Oui =", score_oui)
print("Score Non =", score_non)

# 9. Décision finale
# choisir la plus grande probabilité
if score_oui > score_non:
    print("Classe prédite = Oui")

else:
    print("Classe prédite = Non")
```

## XGBoost (Extreme Gradient Boosting) :

```python
import numpy as np  # pour manipuler facilement les tableaux

# 1. Données : taille (cm), poids (kg) -> âge (années) 
X = np.array([[170, 65], [180, 75], [160, 60]])
y = np.array([100, 200, 300])  # âge cible

# 2. Première prédiction : baseline(commencer avec une valeur simple) = moyenne de y
predictions = np.full_like(y, np.mean(y), dtype=float)

learning_rate = 0.1   # vitesse d’apprentissage
n_estimators = 3  # nombre d'arbres 

# 3. Boucle XGBoost
# chaque arbre : essaie de corriger les erreurs du modèle précédent

for tree_idx in range(n_estimators):
    
    residuals = y - predictions   # Calcul des résidus (erreurs) = vraie valeur - prédiction actuelle
    new_tree_pred = np.mean(residuals)  # correction constante pour tous

# 4. Mise à jour des prédictions : ajouter la correction pondérée
    predictions += learning_rate * new_tree_pred   # nouvelle prédiction = ancienne prédiction + petite correction

# 5. Afficher l'évolution
    print(f"Arbre {tree_idx+1}: résidus = {residuals}, correction = {new_tree_pred:.2f}")
    print(f"  Nouvelles prédictions = {predictions}\n")

# 6. Résultat final
print("Prédictions finales:", predictions)
print("Erreurs finales:", y - predictions)  # erreurs finales
```
