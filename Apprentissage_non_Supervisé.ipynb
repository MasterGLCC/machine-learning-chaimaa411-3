## DBSCAN (Density-Based Spatial Clustering of Applications with Noise) : 

```python
# 1. Données (x , y)
points = [
    (1, 1),
    (1, 2),
    (2, 1),
    (8, 8),
    (8, 9),
    (25, 25)
]
eps = 2  # Distance maximale entre deux voisins
minPts = 2  # Nombre minimum de voisins pour crée une zone dense 
visited = []  # Points déjà visités
clusters = []  # Liste des clusters
noise = []  # Points isolés (bruit)

# 2. Distance euclidienne : Calcul distance euclidienne entre 2 points
def distance(p1, p2):
    x1, y1 = p1  # coordonnées du premier point
    x2, y2 = p2  # coordonnées du deuxième point
    
    return ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5  # formule de distance

# 3. Trouver les voisins d'un point 
def get_neighbors(point):
    neighbors = []   # liste des voisins

    # parcourir tous les points
    for p in points:
        if distance(point, p) <= eps:  # si distance <= eps , alors c’est un voisin
            neighbors.append(p)

    return neighbors

# 4. DBSCAN
for point in points:  # parcourir tous les points
    
    if point in visited:   # ignorer si déjà visité
        continue
    visited.append(point)  # marquer comme visité

    neighbors = get_neighbors(point)   # chercher voisins

    # CAS 1 : pas assez dense => bruit
    if len(neighbors) < minPts:

        noise.append(point)

    #CAS 2 : assez de voisins => nouveau cluster
    else:
        cluster = []   # créer nouveau cluster
        cluster.append(point)  # ajouter point actuel

        # explorer les voisins
        for neighbor in neighbors:

            if neighbor not in visited:   # si voisin pas visité
                visited.append(neighbor)  # marquer visité
                new_neighbors = get_neighbors(neighbor)  # chercher voisins du voisin

                if len(new_neighbors) >= minPts:   # si le voisin est aussi dense

                    for n in new_neighbors:  # ajouter ses voisins
                        if n not in neighbors:
                            neighbors.append(n)

            # ajouter au cluster
            if neighbor not in cluster:
                cluster.append(neighbor)

        clusters.append(cluster)  # ajouter cluster final

# 5. Résultat
print("Clusters :")
for c in clusters:
    print(c)

print("\nBruit :")
print(noise)
```

## PCA (Principal Component Analysis) : 

```python
import math   # pour utiliser cos, sin, atan2, pi

# 1. Données [taille, poids]
data = [

    [170, 65],
    [180, 75],
    [160, 60]
]

# 2. Centrage des données
n = len(data)   # nombre de lignes
mean = [0, 0]  # mean[0] = moyenne des tailles, mean[1] = moyenne des poids

for p in data:  # parcourir les données

    mean[0] += p[0]   # ajouter les tailles
    mean[1] += p[1]   # ajouter les poids

# calcul moyenne finale
mean[0] /= n
mean[1] /= n

# données centrées
centered = [[p[0] - mean[0], p[1] - mean[1]] for p in data]   # nouvelle valeur = ancienne valeur - moyenne

# 3. Matrice de covariance
var_x = sum(p[0]**2 for p in centered) / (n-1)   # variance de X, mesure dispersion des tailles

var_y = sum(p[1]**2 for p in centered) / (n-1)   # variance de Y, mesure dispersion des poids

cov_xy = sum(p[0]*p[1] for p in centered) / (n-1)  # covariance X,Y mesure relation entre taille et poids

print(f"Matrice de covariance :\n[{var_x:.2f}, {cov_xy:.2f}\n {cov_xy:.2f}, {var_y:.2f}]")   # affichage matrice covariance

# 4. Première composante principale
# Calcul du vecteur propre associé à la plus grande valeur propre (première composante)
if var_x != var_y:
    theta = 0.5 * math.atan2(2 * cov_xy, var_x - var_y)    # tan(2θ) = 2*cov_xy / (var_x - var_y)
else:
    theta = math.pi / 4  # cas particulier

# 5. Vecteur directeur
direction = [math.cos(theta), math.sin(theta)]   # Vecteur directeur unitaire (composante principale)

print(f"Première composante (vecteur unitaire) : [{direction[0]:.3f}, {direction[1]:.3f}]")  # afficher composante principale

# 6. Projection des données centrées sur ce vecteur
projected = [p[0]*direction[0] + p[1]*direction[1] for p in centered]

# 7. Résultat final
print("\nDonnées après projection sur la première composante :")
for val in projected:
    print(f"{val:.2f}")
```
